```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
flight_data = pd.read_csv(r'C:\Users\Deepa\Desktop\Python\Data\data-for-datavis\flight_delays.csv', 
                         index_col='Month')
```

```python
flight_data.head()
```

```python
plt.figure(figsize=(16,6))

plt.title('HeatMap to visualize patterns in flight data')

sns.heatmap(data=flight_data, annot=True)

#annot=True - This ensures that the values for each cell appear on the chart

plt.xlabel('Airline')
```

### A closer look on the above data suggests for months 9-11 the flight delays is minimal

```python
ign_data = pd.read_csv(r'C:\Users\Deepa\Desktop\Python\Data\data-for-datavis\ign_scores.csv',
                      index_col='Platform')
```

```python
#rounding off the column values to two places
ign_data = ign_data.round(2)

ign_data.head()
```

```python
sns.set(font_scale=2)

plt.figure(figsize=(18,12))

plt.title('HeatMap to visualize average game score by platform and genre',
         fontsize='30')

sns.heatmap(data=ign_data, annot=True)

#annot=True - This ensures that the values for each cell appear on the chart

plt.xlabel('Genre', fontsize='25')

# plt.tick_params(labelsize = 20)
```
