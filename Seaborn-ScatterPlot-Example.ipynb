```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
insurance = pd.read_csv(r'C:\Users\Deepa\Desktop\Python\Data\data-for-datavis\insurance.csv')
```

```python
insurance.head()
```

```python
sns.scatterplot(x=insurance['bmi'] , 
               y=insurance['charges'])
```

```python
sns.regplot(x=insurance['bmi'] , 
               y=insurance['charges'])
```

```python
sns.scatterplot(x=insurance['bmi'] , 
               y=insurance['charges'],
               hue=insurance['smoker'])
```

```python
#note that for lmplot we can't specify the x and y data as a list i.e. as above 
#we can't have them as df['column_anme']

sns.set(font_scale=1)

sns.lmplot(x='bmi' , 
               y='charges',
               hue='smoker', data=insurance)
```

```python
sns.swarmplot(x=insurance['smoker'] , 
               y=insurance['charges'])
```

```python
candy = pd.read_csv(r'C:\Users\Deepa\Desktop\Python\Data\data-for-datavis\candy.csv', index_col='id')
```

```python
candy = candy.round(2)

candy.head()
```

```python
#most popular among all

(candy.loc[candy['winpercent']==candy['winpercent'].max(),
           ['competitorname','winpercent']])
```

```python
#most sugar among all

(candy.loc[candy['sugarpercent']==candy['sugarpercent'].max(),
          ['competitorname','sugarpercent']])
```

```python
sns.scatterplot(x=candy['sugarpercent'], y=candy['winpercent'])
```

```python
sns.regplot(x=candy['sugarpercent'], y=candy['winpercent'])
```

```python
sns.scatterplot(x=candy['pricepercent'], y=candy['winpercent']
               ,hue=candy['chocolate'])
```

```python
sns.lmplot(x='pricepercent', y='winpercent'
               ,hue='chocolate', data=candy)
```

#### From the above pattern we can say that candies which contain chocolate are popular even when the price is high where as candies which don't contain chocolate are popular only when they are cheap

```python
sns.swarmplot(x=candy['chocolate'], y=candy['winpercent'])
```

### When we have to plot a scatter plot for categorical data use swarm plot for the same as the regular scatter plot expects a continous variable.

### Also as can be seen from the above plot we can easily say that candies with chocolate are more popular than candies without them
